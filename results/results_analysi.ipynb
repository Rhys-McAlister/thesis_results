```python
import pandas as pd

xtb_pre_trained_results = pd.read_csv(r"C:\Users\rhys-\OneDrive\Documents\Github\thesis_results\results\xtb_pre_trained_results.csv")
qm9s_pre_trained_results = pd.read_csv(r"C:\Users\rhys-\OneDrive\Documents\Github\thesis_results\results\qm9s_pre_trained_results.csv")
                                      
xtb_fine_tuned_results = pd.read_csv(r"C:\Users\rhys-\OneDrive\Documents\Github\thesis_results\results\xtb_fine_tuned_results.csv")
qm9s_fine_tuned_results = pd.read_csv(r"C:\Users\rhys-\OneDrive\Documents\Github\thesis_results\results\qm9s_fine_tuned_results.csv") 

qm9s_fine_tuned_ensemble_results = pd.read_csv(r"C:\Users\rhys-\OneDrive\Documents\Github\thesis_results\results\ft_ensemble_results_qm9s.csv")
```

```python
# plot a histogram of the results
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Set up the matplotlib figure
f, axes = plt.subplots(1, 2, figsize=(15, 5), sharex=True)
sns.despine(left=True)

# Plot a simple histogram with binsize determined automatically
sns.histplot(xtb_pre_trained_results, kde=True, color="b", ax=axes[0])
sns.histplot(qm9s_pre_trained_results, kde=True, color="r", ax=axes[1])

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a figure and axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), dpi=300)

# Plot the histogram for SIS values
ax1.hist(xtb_pre_trained_results['sis'], bins=18, alpha=0.7, label='XTB Pre-trained model', edgecolor='black', linewidth=0.5)
ax1.hist(qm9s_pre_trained_results['sis'], bins=18, alpha=0.7, label='QM9s Pre-trained model', edgecolor='black', linewidth=0.5)
ax1.set_xlabel('SIS', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12)
# ax1.set_title('Histogram of SIS Values', fontsize=14)
ax1.tick_params(axis='both', labelsize=10)
ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
ax1.legend(fontsize=10)

# Plot the histogram for RMSE values
ax2.hist(xtb_pre_trained_results['rmse'], bins=18, alpha=0.7, label='XTB Pre-trained model', edgecolor='black', linewidth=0.5)
ax2.hist(qm9s_pre_trained_results['rmse'], bins=18, alpha=0.7, label='QM9s Pre-trained model', edgecolor='black', linewidth=0.5)
ax2.set_xlabel('RMSE', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
# ax2.set_title('Histogram of RMSE Values', fontsize=14)
ax2.tick_params(axis='both', labelsize=10)
ax2.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
ax2.legend(fontsize=10)

# Adjust the layout and remove top and right spines
fig.tight_layout()
for ax in (ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# Save the figure as a high-quality PNG
plt.savefig('comparison_histogram.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a figure and axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), dpi=300)

# Plot the box and whisker plot for SIS values
ax1.boxplot([xtb_pre_trained_results['sis'], qm9s_pre_trained_results['sis']], labels=['XTB Pre-trained model', 'QM9s Pre-trained model'])
ax1.set_ylabel('SIS', fontsize=12)
ax1.tick_params(axis='both', labelsize=10)
ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Plot the violin plot for SIS values
ax1v = ax1.twinx()
ax1v.violinplot([xtb_pre_trained_results['sis'], qm9s_pre_trained_results['sis']], showmedians=True)
ax1v.set_yticks([])

# Plot the box and whisker plot for RMSE values
ax2.boxplot([xtb_pre_trained_results['rmse'], qm9s_pre_trained_results['rmse']], labels=['XTB Pre-trained model', 'QM9s Pre-trained model'])
ax2.set_ylabel('RMSE', fontsize=12)
ax2.tick_params(axis='both', labelsize=10)
ax2.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Plot the violin plot for RMSE values
ax2v = ax2.twinx()
ax2v.violinplot([xtb_pre_trained_results['rmse'], qm9s_pre_trained_results['rmse']], showmedians=True)
ax2v.set_yticks([])

# Adjust the layout and remove top and right spines
fig.tight_layout()
for ax in (ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# Save the figure as a high-quality PNG
plt.savefig('comparison_plot.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```
