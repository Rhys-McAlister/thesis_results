```python
import pandas as pd
from chemprop import data
```

```python
input_path = r"C:\Users\rhys-\OneDrive\data_hnrs\spectra\nist\nist_exp_spectra_fixed.csv"

nist_df = pd.read_csv(input_path)
```

```python
def get_mol_datapoints(df, smiles_column, target_columns):
    smis = df.loc[:, smiles_column].values
    ys = df.loc[:, target_columns].values
    return [data.MoleculeDatapoint.from_smi(smi, y) for smi, y in zip(smis, ys)]

datapoints = get_mol_datapoints(nist_df, 'smiles', nist_df.columns[1:])

```

```python
from chemprop.data import splitting

nist_train_idx, nist_val_idx, nist_test_idx = splitting.split_data(
    datapoints,
    # num_folds=5,
    split='scaffold_balanced',
)
```

```python
# use indexes to select rows from the dataframe
nist_train = nist_df.iloc[nist_train_idx]
nist_val = nist_df.iloc[nist_val_idx]
nist_test = nist_df.iloc[nist_test_idx]

# combine nist_train and nist_val
nist_train_val = pd.concat([nist_train, nist_val])

# save the dataframes to parquet

# nist_train.to_parquet(r"C:\Users\rhys-\OneDrive\data_hnrs\spectra\nist\nist_train.parquet")
# nist_val.to_parquet(r"C:\Users\rhys-\OneDrive\data_hnrs\spectra\nist\nist_val.parquet")
nist_test.to_parquet(r"C:\Users\rhys-\OneDrive\Documents\Github\thesis_results\data\nist\nist_test.parquet")

nist_train_val.to_parquet(r"C:\Users\rhys-\OneDrive\Documents\Github\thesis_results\data\nist\nist_train.parquet")






```
